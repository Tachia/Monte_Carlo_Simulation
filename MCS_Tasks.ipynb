MONTE CARLO SIMULATION TASK

QUESTION:

Show the trajectory of an electron motion in 2D using Monte Carlo Simulation, and hence determine the angle, φ and the mean free path, Λ for ten (10) collisions.

Given that:

φ = 2πa, a --- [0,1] "random valuable"

Λ = -λ.ln(a)

where: Phi, φ - is the angle between the electron collisions with other particles, a - is Bohr radius, lambda, Λ - is the mean free path, λ - is lambda null which is assumed to be 1, ln(a) - is the natural logarithsm of a in base e.

```python
# The first emission of the electron from the source
# import the necessary libraries
import math
import random
from math import pi

count = 0
N = 1000

while count < 1000:
    
    for num in range(N):    
            a = random.uniform(0,1)
    
    count+=1
# I am using these symbols for: λ - lambda null, Λ - mean free path, φ = angle
λ = -1
φ_1 = 2*pi*a 
Λ_1 = λ*math.log(a) 

print ('Number of Iterations: ', count)
print ("\n")
print ('Bohr Radius, a =', a)
print ("\n")
print ('The Angle in Radians, φ_1 =', φ_1)
print ("\n")
print ('The Mean Free Path, Λ_1 =', Λ_1)
```
